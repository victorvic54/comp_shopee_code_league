```python
border = "============================================="
file_start = "--------------------->"

data = {}

with open('coordinate_result_zzzzzz.txt', 'r') as f:
    sentences = f.read().split("\n")
    
    file_name = []
    real_label = []
    tmp_label = []
    real_coor = []
    tmp_coor = None
    count = []
    defect = []
    overlap = []
    normal = []
    qty_label = []
    tmp_count = None
    tmp_defect = None
    tmp_overlap = None
    tmp_normal = None
    tmp_qty_label = None
    isDefect = False
    isOverlap = False
    
    for i in range(len(sentences)):
        if (file_start in sentences[i]):
            file_name.append(sentences[i].split("/")[-1][:-4])
            if (tmp_label != []):
                real_label.append(tmp_label)
            
            if (tmp_coor != None):
                real_coor.append(tmp_coor)
                
            if (tmp_count != None):
                count.append(tmp_count)
                
            if (tmp_defect != None):
                defect.append(tmp_defect)
                
            if (tmp_overlap != None):
                overlap.append(tmp_overlap)
                
            if (tmp_normal != None):
                normal.append(tmp_normal)
                
            if (tmp_qty_label != None):
                qty_label.append(tmp_qty_label)
                
            continue
            
        if (sentences[i] == border or sentences[i] == ''):
            tmp_label = []
            tmp_coor = []
            tmp_count = 0
            tmp_defect = 0
            tmp_overlap = 0
            tmp_normal = 0
            tmp_qty_label = 0
            continue
        
        if ('count' in sentences[i]):
            tmp_count += 1
            
            label_percent = sentences[i].split(":")
            label = label_percent[0].strip()
            percent = int(label_percent[1].strip()[:-1])
            tmp_label.append((label, percent))
        elif ('defect' in sentences[i]):
            tmp_defect += 1
            
            label_percent = sentences[i].split(":")
            label = label_percent[0].strip()
            percent = int(label_percent[1].strip()[:-1])
            tmp_label.append((label, percent))
            isDefect = True
        elif ('overlap' in sentences[i]):
            tmp_overlap += 1
            
            label_percent = sentences[i].split(":")
            label = label_percent[0].strip()
            percent = int(label_percent[1].strip()[:-1])
            tmp_label.append((label, percent))
            isOverlap = True
        elif ('normal' in sentences[i]):
            tmp_normal += 1
            
            label_percent = sentences[i].split(":")
            label = label_percent[0].strip()
            percent = int(label_percent[1].strip()[:-1])
            tmp_label.append((label, percent))
        elif ('qty_label' in sentences[i]):
            tmp_qty_label += 1
            
            label_percent = sentences[i].split(":")
            label = label_percent[0].strip()
            percent = int(label_percent[1].strip()[:-1])
            tmp_label.append((label, percent))
        else:
            curr = [i.strip() for i in sentences[i].split(",")]
            curr = list(map(int, curr))
            
            if (isDefect):
                tmp_coor.append(('black_spot', curr))
                isDefect = False
                continue
                
            if (isOverlap):
                tmp_coor.append(('double_stack', curr))
                isOverlap = False
                continue
                
    print(tmp_coor)
    if (tmp_label != []):
        real_label.append(tmp_label)
        
    if (tmp_coor != None):
        real_coor.append(tmp_coor)

    if (tmp_count == None):
        tmp_count = 0
    else:
        count.append(tmp_count)

    if (tmp_defect == None):
        tmp_defect = 0
    else:
        defect.append(tmp_defect)

    if (tmp_overlap == None):
        tmp_overlap = 0
    else:
        overlap.append(tmp_overlap)

    if (tmp_normal == None):
        tmp_normal = 0
    else:
        normal.append(tmp_normal)

    if (tmp_qty_label == None):
        tmp_qty_label = 0
    else:
        qty_label.append(tmp_qty_label)
        

    data['img_id'] = file_name
    data['acumen_count_pred'] = normal
    data['handwritten_count_pred'] = normal
    data['anomalies_bbox_pred'] = real_coor
    data['probability_label'] = real_label
    data['count'] = count
    data['defect'] = defect
    data['overlap'] = overlap
    data['normal'] = normal
    data['qty_label'] = qty_label
```

# Data process

Numerical cols

```python
from sklearn.preprocessing import MinMaxScaler

def process_numerical_data(df,numerical_cols):
    sacalar = MinMaxScaler()
    df_numerical = sacalar.fit_transform(df[numerical_cols])
    return df_numerical

df_numerical = process_numerical_data(df,numerical_cols)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

def process_categorical_data(df,categorical_cols):
    df_categorical = None
    drop_cols = ['EmployeeCount', 'Over18', 'StandardHours']
    
    for col in [var for var in categorical_cols if var not in drop_cols]:
        lbe = LabelEncoder()
        ohe = OneHotEncoder()

        lbe.fit(df[col].values.reshape(-1, 1))
        df[col] = lbe.transform(df[col])

        ohe.fit(df[col].values.reshape(-1, 1))
        oht_df = ohe.transform(df[col].values.reshape(-1, 1)).todense()

        if df_categorical is None:
            df_categorical = oht_df
        else:
            df_categorical = np.hstack((df_categorical, oht_df))
    return df_categorical
```

```python
df_categorical = process_categorical_data(df,categorical_cols)
df_categorical
```

```python
df_features = np.hstack((df_numerical, df_categorical))
df_features.shape
```

Labels

```python
def process_label(df):
    target_col_dict = {'Yes': 1, 'No': 0}
    df_labels = df[target_col].map(target_col_dict).values
    return df_labels
df_labels = process_label(df)
df_labels.shape
```

# Set Up Model

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_features,df_labels, test_size=0.2, random_state=23)
```

```python
from sklearn import metrics
from sklearn.ensemble import AdaBoostClassifier, ExtraTreesClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.tree import DecisionTreeClassifier

import xgboost as xgb
```

```python
# Weird solution but give a high prediction:
# KFold and only grab the highest prediction from the batch
# In general: KFold should only be used as validation

rng = np.random.RandomState(31337)

y = df_labels
X = df_features
THRESHOLD = 0.94

kf = KFold(n_splits=20, shuffle=True, random_state=rng)
result = []

for train_index, test_index in kf.split(X):
    xgb_model = xgb.XGBClassifier().fit(X[train_index], y[train_index])
    lr_model = LogisticRegression(random_state=42, max_iter=300).fit(X[train_index], y[train_index])
    mlp_model = MLPClassifier(max_iter=800).fit(X[train_index], y[train_index])
    
    predictions_xgb = xgb_model.predict(X[test_index])
    predictions_lr = lr_model.predict(X[test_index])
    predictions_mlp = mlp_model.predict(X[test_index])
    actuals = y[test_index]
    
    if (metrics.accuracy_score(actuals, predictions_xgb) > THRESHOLD):
        result.append(xgb_model)
        
    if (metrics.accuracy_score(actuals, predictions_lr) > THRESHOLD):   
        result.append(lr_model)
        
    if (metrics.accuracy_score(actuals, predictions_mlp) > THRESHOLD):
        result.append(mlp_model)
```

```python
final = []

for classifier in result:
    final.append(pd.DataFrame(classifier.predict(test_x)))
    print(metrics.accuracy_score(test_y, pd.DataFrame(classifier.predict(test_x))))
    
df = pd.concat(final, axis=1)
new_model_kfold = LogisticRegression()
new_model_kfold.fit(df, test_y)
```

```python
NN = MLPClassifier(max_iter=200)
RR = RandomForestClassifier()
MNB = MultinomialNB()

model0 = GradientBoostingClassifier(learning_rate=0.01,random_state=1)
model1 = AdaBoostClassifier(random_state=24)
model2 = KNeighborsClassifier(n_neighbors=8, weights="distance")
model3 = LogisticRegression(random_state=42, max_iter=300)
model4 = ExtraTreesClassifier()
model5 = KNeighborsClassifier(n_neighbors=4)
model6 = KNeighborsClassifier(n_neighbors=16)
model7 = xgb.XGBClassifier(random_state=1, learning_rate=0.01)


NN.fit(train_x,train_y)
RR.fit(train_x, train_y)
MNB.fit(train_x, train_y)

model0.fit(train_x, train_y)
model1.fit(train_x, train_y)
model2.fit(train_x, train_y)
model3.fit(train_x, train_y)
model4.fit(train_x, train_y)
model5.fit(train_x, train_y)
model6.fit(train_x, train_y)
model7.fit(train_x, train_y)

# accuracy calculation

y_pred_0 = pd.DataFrame(NN.predict(test_x))
y_pred_1 = pd.DataFrame(RR.predict(test_x))
y_pred_2 = pd.DataFrame(MNB.predict(test_x))
y_pred_3 = pd.DataFrame(model0.predict(test_x))
y_pred_4 = pd.DataFrame(model1.predict(test_x))
y_pred_5 = pd.DataFrame(model2.predict(test_x))
y_pred_6 = pd.DataFrame(model3.predict(test_x))
y_pred_7 = pd.DataFrame(model4.predict(test_x))
y_pred_8 = pd.DataFrame(model5.predict(test_x))
y_pred_9 = pd.DataFrame(model6.predict(test_x))
y_pred_10 = pd.DataFrame(model7.predict(test_x))


print("Accuracy of NN:", metrics.accuracy_score(test_y, y_pred_0))
print("Accuracy of RR:", metrics.accuracy_score(test_y, y_pred_1))
print("Accuracy of MNB:", metrics.accuracy_score(test_y, y_pred_2))
print("Accuracy of GBM:", metrics.accuracy_score(test_y, y_pred_3))
print("Accuracy of ADA:", metrics.accuracy_score(test_y, y_pred_4))
print("Accuracy of KNN:", metrics.accuracy_score(test_y, y_pred_5))
print("Accuracy of LR:", metrics.accuracy_score(test_y, y_pred_6))
print("Accuracy of EXTRA:", metrics.accuracy_score(test_y, y_pred_7))
print("Accuracy of KNN4:", metrics.accuracy_score(test_y, y_pred_8))
print("Accuracy of KNN16:", metrics.accuracy_score(test_y, y_pred_9))
print("Accuracy of XGB:", metrics.accuracy_score(test_y, y_pred_10))

df = pd.concat([y_pred_0, y_pred_1, y_pred_2, y_pred_3, y_pred_4, y_pred_5, y_pred_6, y_pred_7, y_pred_8, y_pred_9, y_pred_10], axis=1)

new_model = LogisticRegression(random_state=42)
new_model.fit(df, test_y)
```

```python
from statistics import mean
from numpy import std
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score

# evaluate a give model using cross-validation
def evaluate_model(model):
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
    return scores

# get the models to evaluate
models = [NN, RR, MNB, model0, model1, model2, model3, model4, model5, model6, model7]

# evaluate the models and store results
for model in models:
    print(model)
    scores = evaluate_model(model)
    print(scores)
    print('> %.3f (%.3f)' % (mean(scores), std(scores)))
    print("===========================================")
    
    
# ensemble kfold
print(new_model_kfold)
scores = evaluate_model(new_model_kfold)
print(scores)
print('> %.3f (%.3f)' % (mean(scores), std(scores)))
print("===========================================")
    
# ensemble multimodel
print(new_model)
scores = evaluate_model(new_model)
print(scores)
print('> %.3f (%.3f)' % (mean(scores), std(scores)))
print("===========================================")
```

# GridSearch to find the optimal parameter setting

Most used parameter

1）The ith element represents the number of neurons in the ith hidden layer.
    example: hidden_layer_sizes=(10)

2）activation function
    example: activation="relu"

3）The solver for weight optimization. {'lbfgs', 'sgd', 'adam'}
    example: solver='adam'

4）L2 penalty (regularization term) parameter
    example: alpha=0.0001

5）Size of minibatches for stochastic optimizers.
    If the solver is 'lbfgs', the classifier will not use minibatch.
    When set to "auto", `batch_size=min(200, n_samples)`
    example: batch_size='auto'

6）Learning rate schedule for weight updates.
    example: learning_rate="constant"

7）The initial learning rate used. It controls the step-size
    in updating the weights. Only used when solver='sgd' or 'adam'.
    example: learning_rate_init=0.001

8）Maximum number of iterations. The solver iterates until convergence
    example: max_iter=200

9）Tolerance for the optimization. When the loss or score is not improving
    by at least ``tol`` for ``n_iter_no_change`` consecutive iterations,
    unless ``learning_rate`` is set to 'adaptive', convergence is
    considered to be reached and training stops.
    example: tol=1e-4

```python
import time
start_time = time.time()
NN_new = MLPClassifier(max_iter=1000)
parameter_space = {
    'hidden_layer_sizes': [(50,50,50), (50,100,50), (50,100,50,25)],
    'activation': ['tanh', 'relu','logistic'],
    'solver': ['adam', 'lbfgs'],
    'alpha': [0.001, 0.01],
    'learning_rate': ['constant','adaptive'],
}
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(NN_new, parameter_space, n_jobs=-1, cv=10)
clf.fit(train_x, train_y)
print('Time taken for training the model: '+ str(time.time() - start_time))
```

```python
# Best paramete set
print('Best parameters found:\n', clf.best_params_)
```

```python
# All results
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```
